```python
import pandas as pd
import numpy as np
from urllib.request import urlopen
import webdriver_manager
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import sqlite3
```

```python
categories = pd.read_csv("flipkartLinks.txt" , sep=",")
all_links = categories["link"]
all_links
```

```python
start = 60
end = 65
```

```python
final_urls=[]
for index in range(start,end):
    for j in [all_links[index]+str(i) for i in range(2,30)]:
        final_urls.append(j)
```

```python
len(final_urls)
```

```python

driver = webdriver.Chrome(ChromeDriverManager().install())

product_link=[]
product_img_link=[]
product_name=[]
product_price=[]
product_mrp_price=[]
product_brand=[]

for url in final_urls:
    driver.get(url)
    driver.implicitly_wait(50)
    soup=BeautifulSoup(driver.page_source,'html.parser')
    classes=soup.findAll('div',{'class':"_13oc-S"}) 
    
    for product in classes:
            try:
                plink= (product.findAll('a',{'class':"_2UzuFa"}))
                plink = 'https://www.flipkart.com'+plink[0]['href']  if  plink else np.nan
                print(plink)
                product_link.append(plink)
                
                brand=(product.findAll('div',{'class':"_2WkVRV"}))   
                brand = brand[0].text  if brand else np.nan
                print(brand)
                product_brand.append(brand)
                
                name=(product.findAll('a',{'class':"IRpwTa"}))
                name = name[0].text  if name else np.nan 
                print(name)
                product_name.append(name)
                
                price=(product.findAll('div',{'class':"_30jeq3"}))
                price = price[0].text  if price else np.nan
                print(price)
                product_price.append(price)
                
                mrp_price=(product.findAll('div',{'class':"_3I9_wc"}))  
                mrp_price = mrp_price[0].text  if mrp_price else np.nan
                print(mrp_price)
                product_mrp_price.append(mrp_price)
                
                product_img =product.findAll('img',{'class':'_2r_T1I'})
                product_img = product_img[0]['src'].split('?')[0]  if product_img else np.nan
                print(product_img)
                product_img_link.append(product_img)

            except:continue

```

# Creating Dataframe

```python
# Create a base CSV file for storing
columns = ["Website", "Product_Link", "Product_Name", "Product_Brand", "Price",
          "MRP", "Primary_Image_Links"]

product_details_sel = pd.DataFrame(columns=columns, index = None)

product_details_sel['Product_Link']=product_link
product_details_sel['Primary_Image_Links']=product_img_link
product_details_sel['Product_Name']=product_name
product_details_sel['Product_Brand']=product_brand
product_details_sel['Price']=product_price
product_details_sel['MRP']=product_mrp_price
```

```python
product_details_sel["Website"] = "https://www.flipkart.com"
```

```python
product_details_sel.shape
```

```python
product_details_sel.head()
```

```python
product_details_sel.isna().sum()
```

```python
product_details_sel["Price"] = product_details_sel["Price"].str.replace("₹","")
product_details_sel["Price"] = product_details_sel["Price"].str.replace(",","")
product_details_sel["MRP"] = product_details_sel["MRP"].str.replace("₹","")
product_details_sel["MRP"] = product_details_sel["MRP"].str.replace(",","")
```

```python
product_details_sel.drop_duplicates(inplace = True)
product_details_sel.dropna(inplace = True)
product_details_sel.to_csv('product_details1.csv',index=False)
```

# DataBase

```python
conn = sqlite3.connect('flipkartDb1.db')
c = conn.cursor()
```

```python
c.execute('CREATE TABLE IF NOT EXISTS product_details (Website varchar(40) , Product_Link TEXT PRIMARY KEY,Product_Name varchar(50) NOT NULL,Product_Brand varchar(50) NOT NULL,Product_Category varchar(50),Size_Avail varchar(20) ,Price int NOT NULL,MRP int NOT NULL,Gender varchar(15) ,Description TEXT ,Primary_Image_Links TEXT NOT NULL,Secondary_Image_Links TEXT ,Affiliate_Link TEXT)')
conn.commit()
```

```python
product_details_sel.to_sql('product_details', conn, if_exists='replace', index = False)
```
